```python
import os
import boto3

print("Libraries imported successfully!")
```

```python
# 1. Get the connection details from the environment
endpoint = os.getenv("AWS_S3_ENDPOINT")
access_key = os.getenv("AWS_ACCESS_KEY_ID")
secret_key = os.getenv("AWS_SECRET_ACCESS_KEY")
bucket_name = os.getenv("AWS_S3_BUCKET")

# 2. Create the "S3 Client" (The bridge to MinIO)
s3_client = boto3.client(
    's3',
    endpoint_url=endpoint,
    aws_access_key_id=access_key,
    aws_secret_access_key=secret_key,
    verify=False  # This ignores SSL errors common in lab environments
)

print(f"Connection bridge created for bucket: {bucket_name}")
```

```python
# The name of the file on your workbench
local_file = 'test.txt'

# The name you want the file to have inside the bucket
destination_name = 'test.txt'

try:
    s3_client.upload_file(local_file, bucket_name, destination_name)
    print(f"✅ Success! {local_file} is now inside {bucket_name}")
except FileNotFoundError:
    print("❌ Error: I couldn't find 'test.txt'. Make sure it is in the same folder as this notebook!")
except Exception as e:
    print(f"❌ An error occurred: {e}")
```
